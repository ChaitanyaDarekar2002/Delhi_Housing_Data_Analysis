```python
data=data.drop_duplicates()
```

```python
data.columns
```

```python
data['Balcony']=data['Balcony'].fillna('0')
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.columns
```

```python
data['Status']=data['Status'].fillna('Not Define')
```

```python
data['parking']=data['parking'].fillna('0')
```

```python
data['Furnished_status']=data['Furnished_status'].fillna('Not Define')
```

```python
data['Lift']=data['Lift'].fillna('0')
```

```python
data['Landmarks']=data['Landmarks'].fillna('Not Define')
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.hist(data['price'],bins=30)
plt.show()
```

```python
data['Bedrooms'].value_counts().plot(kind='bar')
plt.title('Bedrooms_Count')
```

```python
Furnishing=data['Furnished_status'].value_counts()
plt.pie(
    Furnishing,
    labels=Furnishing.index,
    autopct='%1.1f%%'
)
plt.title("Furnishing Distribution")
plt.show()
```

```python
Lifts=data['Lift'].value_counts()
plt.pie(Lifts,labels=Lifts.index,autopct='%1.1f%%')
plt.title('lift_Count')
```

```python
plt.scatter(data['Bedrooms'],data['price'])
```

```python
top_locations=data['area'].value_counts().head(10)

plt.figure(figsize=(12,6))

top_locations.plot(kind='bar',color='red')

plt.title("Top 10 Localities")
plt.xlabel("Locality")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.scatter(data['area'],data['price'],color='red')
```

```python
numeric=data.select_dtypes(include='number')

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric.corr(),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```
